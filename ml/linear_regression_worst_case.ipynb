```python
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# define data X and y arrays
"""
age_x = [4,6,7,6,1,16,1,8,3,10,11,8,5]
capacity_y = [98,85,86,89,112,85,101,84,93,79,76,82,81]
"""
"""
# data with correct fit
age_x = [4,6,7,6,1,16,1,8,3,10,11,8,5]
capacity_y = [14,35,48,36,1,254,2,63,9,99,120,62,25]
"""

# data with worst fit
age_x = [4,6,7,6,1,16,1,8,3,10,11,8,5]
capacity_y = [63,9,99,62,25,14,35,48,36,1,124,2,63]

slope, intercept, r, p, std_error = stats.linregress(age_x, capacity_y)
print (f"slope = {slope}, intercept = {intercept}, r = {r}, p = {p} and std_error = {std_error}")
```

```python
def new_y(x):
    return slope * x + intercept
```

```python
regressed_values = list(map(new_y, age_x))

plt.scatter(age_x, capacity_y, label="Original data")
plt.plot(age_x, regressed_values, "r", label="Fitted Line")
plt.xlabel("Age of the machine")
plt.ylabel("Capacity of machine")
plt.legend()
plt.show()
```
